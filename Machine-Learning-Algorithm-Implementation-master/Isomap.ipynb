```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.manifold import Isomap
```

##### Reading data

```python
digits = load_digits()
```

```python
dir(digits)
```

##### algorithm

```python
isom = Isomap(n_components=2)
predict = isom.fit_transform(digits.data)
```

```python
print(digits.data.shape)
print(predict.shape)
```

##### plotting

```python
plt.scatter(predict[:,0], predict[:,1], c=digits.target)
plt.colorbar()
```
