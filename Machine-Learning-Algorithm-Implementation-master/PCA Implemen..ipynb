```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
```

##### Reading data

```python
digits = load_digits()
dir(digits)
```

```python
print(digits.images.shape)
print(digits.data.shape)
print(digits.target.shape)
```

##### algorithm

```python
pca = PCA(n_components=2)
predict = pca.fit_transform(digits.data)
```

##### plotting

```python
plt.scatter(predict[:,0], predict[:,1], c=digits.target)

plt.xlabel('predict 1')
plt.ylabel('predict 2')
plt.colorbar()
```
