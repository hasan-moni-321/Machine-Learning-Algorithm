```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
```

##### reading data

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data.shape
```

```python
img, axes = plt.subplots(2,3, figsize=(8,8), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=.1, wspace=.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i].reshape(8,8), interpolation='nearest')
```

##### algorithm

```python
mds = MDS(n_components=3)
transform = mds.fit_transform(digits.data)
```

```python
print(transform.shape)
print(digits.data.shape)
```

##### plotting

```python
plt.scatter(transform[:,0], transform[:,1], c=digits.target)
plt.colorbar()
```
