```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import FastICA
```

##### Reading data

```python
faces = fetch_olivetti_faces()
```

```python
dir(faces)
```

```python
print(faces.images.shape)
print(faces.data.shape)
print(faces.target.shape)
```

```python
fig, axes = plt.subplots(2,3, figsize=(10,4), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=.1, wspace=.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(faces.images[i].reshape(64,64),interpolation='nearest')
```

##### algorithm

```python
fica = FastICA(n_components=2)
transformed_data = fica.fit_transform(faces.data)
```

```python
print(transformed_data.shape)
print(faces.data.shape)
```

##### plotting

```python
plt.scatter(transformed_data[:,0],transformed_data[:,1], c=faces.target)
plt.colorbar()
```
